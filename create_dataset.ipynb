```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
```

```python
covid_df = pd.read_csv('xRay_Corona/metadata.csv')
covid_df.head()
```

```python
covid_df.groupby('finding').count()
```

```python
covid_df = covid_df[lambda x: x['view'] == 'PA']
covid_df.shape
```

```python
covid_df = covid_df[covid_df['finding'].isin(['COVID-19', 'COVID-19, ARDS'])]
covid_df.shape
```

```python
covid_df = covid_df[['folder', 'filename']]
covid_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test, _, __ = train_test_split(covid_df['filename'], covid_df['folder'], test_size = 0.2)
```

```python
train.shape
```

```python
test.shape
```

```python
or_path = 'xRay_Corona/images'
import shutil
```

```python
for i in train:
    shutil.copyfile(or_path+'/'+i, 'data/train/positive/'+i)
```

```python
for i in test:
    shutil.copyfile(or_path+'/'+i, 'data/test/positive/'+i)
```
